```python
from collections import deque

def BFS(puzzle):
    rows = len(puzzle)
    cols = len(puzzle[0])
    visited = [[False] * cols for _ in range(rows)]  # 记录已访问的位置
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]  # 右、左、下、上
    start = None  # 起点位置
    end = None  # 终点位置

    # 找到起点和终点的位置
    for i in range(rows):
        for j in range(cols):
            if puzzle[i][j] == 'S':
                start = (i, j)
            elif puzzle[i][j] == 'E':
                end = (i, j)

    path = []  # 记录移动过程中的位置信息
    queue = deque([(start[0], start[1], [])])  # 使用队列来进行广度优先搜索

    while queue:
        row, col, curr_path = queue.popleft()

        if row < 0 or row >= rows or col < 0 or col >= cols or puzzle[row][col] == '1' or visited[row][col]:
            continue

        visited[row][col] = True
        curr_path.append((row, col))
        print(curr_path)
        if (row, col) == end:
            path = curr_path
            break
            
        for dx, dy in directions:
            queue.append((row + dx, col + dy, curr_path[:]))  # 将当前路径的副本添加到队列中

    return path
```
